```python
%matplotlib inline
import math
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
# import plotly.plotly as py
# import plotly.graph_objs as go
from sympy.mpmath import *
# plotly.plotly.sign_in("joknowle", "h1w8dn6q7h")
matplotlib.rcParams.update({'font.size': 12, 'font.family': 'calibri', 'mathtext.fontset': 'stix'})
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
n = np.linspace(0,50,51); N = 50; multiplicity = np.zeros(51);

for i, x in enumerate(n):
    multiplicity[i] = fac(N)/(fac(n[i])*(fac(N-n[i])))

a, = plt.plot(n, multiplicity, 'r', label='Probability Curve', linewidth=2);


plt.title('Probability of Getting Heads as a function of n')
plt.xlabel('n, # of coin heads selected')
plt.ylabel('Probability')
plt.legend(handles=[a])
plt.show();
```

```python
q = np.linspace(0,2,3); N = 4; multiplicity = np.zeros(4);
#print(multiplicity)
for i, x in enumerate(q):
    multiplicity[i] = fac(q[i]+N-1)/(fac(q[i])*(fac(N-1)))
print (max(multiplicity))
```

```python
q = np.linspace(0,100,101); NA = 200; NB = 100; mA = np.zeros(200); mB = np.zeros(200); seq = np.zeros(5); results = np.zeros((101,5))
#print(results)
for i, x in enumerate(q):
    mA[i] = fac(q[i]+NA-1)/(fac(q[i])*(fac(NA-1)));
    mB[-i-1] = fac(q[-i-1]+NB-1)/(fac(q[-i-1])*(fac(NB-1)))
    results[i] = [q[i], mA[i], q[-i-1], mB[-i-1], mA[i]*mB[-i-1]]
print (results)
```

```python
#print(results[:,[1]])
fig, ax = plt.subplots(figsize=(12, 8))
a, = plt.plot(results[:,[0]], results[:,[4]], 'r', label='Distribution of Energy in Oscillator A', linewidth=2);
b, = plt.plot(results[:,[2]], results[:,[4]], 'b', label='Distribution of Energy in Oscillator B', linewidth=2);
plt.title('sdf')
plt.xlabel('Distribution of Energy')
plt.axis([0, 100, 0, 2e95])
plt.ylabel('Total Multiplicity')
plt.legend(handles=[a,b])
plt.show();
```

```python
print('Most probable state: ', max(results[:,[4]]), 'where qA =', np.argmax(results[:,[4]]))
print('The likelyhood of this happening is ', max(results[:,[4]])/1.68e96*100, '%')
```

```python
print('Least probable state: ', min(results[:,[4]]), 'where qA =', np.argmin(results[:,[4]]))
print('The likelyhood of this happening is ', min(results[:,[4]])/1.68e96*100, '%')
```
